```python
# Import Numpy & PyTorch
import numpy as np
import torch
```

**Training Data**

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43],
                   [91, 88, 64],
                   [87, 134, 58],
                   [102, 43, 37],
                   [69, 96, 70]], dtype = 'float32')
```

```python
# Target (apples)
targets = np.array([[56],
                    [81],
                    [119],
                    [22],
                    [103]], dtype='float32')
```

```python
# Convert inputs and targets to tensors
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)

print(inputs)
print(targets)
```

**Linear Regression Model(from Scratch)**

```python
# Weights and biases
w = torch.randn(3, 1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

print(w)
print(b)
```

```python
# Define the model
def model(x):
    return x @ w + b
```

```python
# Generate predictions
preds = model(inputs)
print(preds)
```

```python
# Compare with targets
print(targets)
```

**Loss Function**

```python
def mse(t1, t2):
    diff = t1 - t2
    return torch.sum(diff * diff) / diff.numel()
```

```python
loss = mse(preds, targets)
print(loss)
```

**Compute Gradients**

```python
# Compute gradients
loss.backward()
```

```python
# Gradients for weights
print(w)
print(w.grad)
```

```python
# Gradients for bias
print(b)
print(b.grad)
```

```python
w.grad.zero_()
b.grad.zero_()
print(w.grad)
print(b.grad)
```

**Adjust weights and bias using gradient descent**

```python
# Generate predictions
preds = model(inputs)
print(preds)
```

```python
# Calculate the loss
loss = mse(preds, targets)
print(loss)
```

```python
# Compute gradients
print(w.grad)
print(b.grad)
```

```python
# Adjust weights and reset gradients
with torch.no_grad():
    w -= w.grad * 1e-5
    b -= b.grad * 1e-5
    w.grad.zero_()
    b.grad.zero_()
```

```python
print(w)
print(b)
```

```python
# Calculate loss
preds = model(inputs)
loss = mse(preds, targets)
print(loss)
```

**Train for multiple epochs**

```python
# Train for 100 epochs
for i in range(100):
    preds = model(inputs)
    loss = mse(preds, targets)
    loss.backward()
    with torch.no_grad():
        w -= w.grad * 1e-5
        b -= b.grad * 1e-5
        w.grad.zero_()
        b.grad.zero_()
```

```python
# Calculate loss
preds = model(inputs)
loss = mse(preds, targets)
print(loss)
```
